## Getting useful columns

```python
even_filtered = even[['developerResponse','rating','review','title']]
even_filtered.head()
```

## Removing digits from 'review' column

```python
def remove_digits(comment):
    return ''.join([ch for ch in comment if not ch.isdigit()])

even_filtered['review'] = even_filtered['review'].apply(remove_digits)
even_filtered.head()
```

## Tokenizing

```python
punctuations = string.punctuation

# Create our list of stopwords
nlp = spacy.load('en')
stop_words = spacy.lang.en.stop_words.STOP_WORDS

# Load English tokenizer, tagger, parser, NER and word vectors
parser = English()

# Creating our tokenizer function
def spacy_tokenizer(sentence):
    # Creating our token object, which is used to create documents with linguistic annotations.
    mytokens = parser(sentence)

    # Lemmatizing each token and converting each token into lowercase
    mytokens = [ word.lemma_.lower().strip() if word.lemma_ != "-PRON-" else word.lower_ for word in mytokens ]

    # Removing stop words
    mytokens = [ word for word in mytokens if word not in stop_words and word not in punctuations ]

    # return preprocessed list of tokens
    return mytokens
```

## Count vectorizer DTM

```python
corpus = np.array(list(even_filtered['review']))
countvec = CountVectorizer(tokenizer = spacy_tokenizer, ngram_range=(1,1))
count = countvec.fit_transform(corpus)
dtm_count = pd.DataFrame(count.toarray(), columns=countvec.get_feature_names(), index=even_filtered.index)
```
