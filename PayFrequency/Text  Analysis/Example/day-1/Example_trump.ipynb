## Opening the dataset

```python
import os
import pandas as pd
DATA_DIR = 'data'

fname = 'trump-tweets.csv'
fname = os.path.join(DATA_DIR, fname)
df = pd.read_csv(fname) 
df
```

## Document-term matrix

```python
from sklearn.feature_extraction.text import CountVectorizer
countvec = CountVectorizer(min_df=2, stop_words='english', binary=True, ngram_range=(2,2))
dtm = pd.DataFrame(countvec.fit_transform(df.Tweet_Text).toarray(), columns=countvec.get_feature_names(), index = df.index) #turns the matrix baxk into a readable dataset 
dtm.head(10)
```
